```python
#Question 3:

import numpy as np
import numpy.linalg as la

#Q-3: a)

def vandermonde(N):
    vec = np.arange(N) + 1  
    vander_matrix = vec[:, np.newaxis] ** np.arange(N)
    return vander_matrix

vander = vandermonde(12)
print("Vandermonde matrix for N = 12:\n", vander)

#Q-3: b)

# Create a vector of ones with length 12
x = np.ones(12)

# Perform matrix-vector multiplication
b = np.dot(vander, x)
print("\nVector b after matrix-vector multiplication:\n", b)

#Q-3: c)

# Compute the inverse of the Vandermonde matrix
vander_inv = np.linalg.inv(vander)

v = np.linalg.inv(vander)

# Multiply the inverse matrix by vector b to solve for x
x_naive = np.dot(vander_inv, b)
print("\nNaïve solution (using inverse of Vandermonde matrix):\n", x_naive)

#Q-3: d)

# Solve the system using numpy.linalg.solve
x_solve = np.linalg.solve(vander, b)
print("\nSolution using numpy.linalg.solve:\n", x_solve)
```
